```python
import pandas as pd
import numpy as np
from src.similarity import consultar_item, treinar_knn
import warnings

warnings.filterwarnings("ignore")
```

### Carregar dados gerados anteriormente ###

```python
print('Carregando embeddings...')
df_embeddings = pd.read_parquet('data/staging/embeddings_bp.parquet')
df_embeddings.head(5)
```

```python
print('Carregando a base padronizada para correlação...')
df_pad = pd.read_csv('data/staging/df_pad.csv')
df_pad.head(5)
```

### Enviar embeddings para o Modelo ###

```python
# Enviar os embeddings gerados para o KNN
knn_model = treinar_knn(df_embeddings, metric='cosine', n_neighbors=30)
knn_model
```

### Busca por item similar ###

```python
# Entrada do usuário
insumo_consulta = input('Insira a descrição completa do insumo: (Ex: ARROZ POLIDO TIPO 1 LONGO FINO PACOTE) ')
marca_consulta = input('Considere a MARCA caso queira uma priorização de marca: (Ex: TIO JOAO, COMBRASIL, CARRETEIRO,...) ')
medida_consulta = input('Considere a MEDIDA caso queira uma priorização de medida: (Ex: 1KG, 500G, 750ML, 50MG,...) ')
```

```python
# Busca por similaridade
resultados = consultar_item(insumo_consulta, medida_consulta, knn_model, df_pad, df_embeddings, marca_consulta)
```
